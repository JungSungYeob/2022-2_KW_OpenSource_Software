GrayScale로 이미지 띄우고 histogram 보이기

```python
import cv2
import numpy as np
import matplotlib.pylab as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/jpg/img1027_jpg.rf.056bd301d096cf5287150b6394a3f4d9.jpg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)

hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)

print("hist.shape:", hist.shape)
print("hist.sum():", hist.sum(), "img.shape", img.shape)
plt.show()
```

```python
img = cv2.imread('/content/drive/MyDrive/jpg/img1027_jpg.rf.056bd301d096cf5287150b6394a3f4d9.jpg')
cv2_imshow(img)

channels = cv2.split(img)
colors = ('b', 'g', 'r')
for(ch, color) in zip (channels, colors):
  hist = cv2.calcHist([ch], [0], None, [256], [0,256])
  plt.plot(hist, color = color)
plt.show()
```

histo_equalize

```python
img = cv2.imread('/content/drive/MyDrive/jpg/img1027_jpg.rf.056bd301d096cf5287150b6394a3f4d9.jpg')
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
img2 = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

cv2_imshow(img)
cv2_imshow(img2)
```

Contrast linited Adaptive histogram Equalization

```python
img = cv2.imread('/content/drive/MyDrive/jpg/img829_jpg.rf.f03f5f137cc0791a3e9438f54cc11255.jpg')
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

img_eq = img_yuv.copy()
img_eq[:,:,0] = cv2.equalizeHist(img_eq[:,:,0])
img_eq = cv2.cvtColor(img_eq, cv2.COLOR_YUV2BGR)
img_clahe = img_yuv.copy()
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
img_clahe[:,:,0] = clahe.apply(img_clahe[:,:,0])
img_clahe = cv2.cvtColor(img_clahe, cv2.COLOR_YUV2BGR)

cv2_imshow(img)
cv2_imshow(img_clahe)
cv2_imshow(img_eq)

channels = cv2.split(img)
colors = ('b', 'g', 'r')
for(ch, color) in zip (channels, colors):
  hist = cv2.calcHist([ch], [0], None, [256], [0,256])
  plt.plot(hist, color = color)
plt.show()

channels = cv2.split(img_clahe)
colors = ('b', 'g', 'r')
for(ch, color) in zip (channels, colors):
  hist = cv2.calcHist([ch], [0], None, [256], [0,256])
  plt.plot(hist, color = color)
plt.show()
```
